```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from googletrans import Translator

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('unigram_freq.csv')

# Create a list of stop words
stop_words = set(stopwords.words('english'))

# Remove stop words from the DataFrame
df = df[~df['word'].isin(stop_words)]

# Sort the DataFrame by count in descending order
df = df.sort_values(by='count', ascending=False)

# Take the 50 most frequent words
df = df.head(50)

```

```python
df
```

```python

# Create a Translator object
translator = Translator()

# Translate the words to Bangla
df['Bangla'] = df['word'].apply(lambda x: translator.translate(x, dest='bn').text)

# Rename the "word" column to "English"
df = df.rename(columns={'word': 'English'})

# Reorder the columns
df = df[['English', 'Bangla', 'count']]

```

```python
df.shape
```

```python
df = df.drop(df[df['English'] == df['Bangla']].index)
```
