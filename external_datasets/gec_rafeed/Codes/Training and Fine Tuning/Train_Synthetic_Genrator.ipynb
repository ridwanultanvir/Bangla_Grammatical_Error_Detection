```python
# Imports
from happytransformer import HappyTextToText
from happytransformer import TTSettings
from happytransformer import TTTrainArgs
import pandas as pd
import torch
import gc
```

```python
# Initializing
parent_path = '/content/drive/MyDrive/Google Colab/Advanced NLP'
T5 = HappyTextToText("T5", "vennify/t5-base-grammar-correction")
```

```python
# Testing if it works
beam_settings =  TTSettings(num_beams=5, min_length=1, max_length=20)
result = T5.generate_text("grammar: I boughts ten apple.", args=beam_settings)
print(result.text)
```

```python
# Loading training dataset
lang8 = '/content/drive/MyDrive/Google Colab/Advanced NLP/Hannah_Datasets/lang8/lang8.csv'
# df = pd.read_csv(lang8)
```

```python
modelPath = '/content/drive/MyDrive/Google Colab/Advanced NLP/Syn_Model/'
T5 = HappyTextToText(load_path=modelPath)

for epoch in range(1):
  args = TTTrainArgs(batch_size=8, learning_rate=0.0001, num_train_epochs=1, max_input_length=256, max_output_length=256)
  T5.train(lang8, args=args)
  T5.save(modelPath)
```
