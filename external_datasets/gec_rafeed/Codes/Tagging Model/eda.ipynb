```python
import pandas as pd
import numpy as np
from datasets import load_dataset
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('../data/lang-8/train_tagged.parquet')
```

```python
labels = df.label.to_numpy()
```

```python
classes = ['ADJ', 
           'ADJ:FORM', 
           'ADV', 
           'CONJ', 
           'CONTR', 
           'DET', 
           'MORPH', 
           'NOUN', 
           'NOUN:INFL',
           'NOUN:NUM', 
           'NOUN:POSS', 
           'ORTH', 
           'OTHER', 
           'PART', 
           'PREP', 
           'PRON', 
           'PUNCT', 
           'SPELL', 
           'VERB', 
           'VERB:FORM', 
           'VERB:INFL', 
           'VERB:SVA',
           'VERB:TENSE', 
           'WO']
```

```python
labels = np.concatenate(labels).reshape(-1,len(classes))
```

```python
labels_df = pd.DataFrame(labels, columns=classes)
```

```python
labels_df.sum()
```

```python
sns.set_theme(style='white', rc = {'figure.figsize':(6,4)})
sns.histplot(data=labels_df.sum())
plt.xlabel('# of Entries with Label')
plt.ylabel('# of Labels')
plt.show()
```

```python
sns.set_theme(style='white', rc = {'figure.figsize':(20,8)})
sns.barplot(x=labels_df.columns, y=labels_df.sum().values, palette='tab20_r', log=True)
plt.xticks(rotation=30, ha='right')
plt.xlabel('Label Name')
plt.ylabel('# of Entries with Label')
plt.show()
```
